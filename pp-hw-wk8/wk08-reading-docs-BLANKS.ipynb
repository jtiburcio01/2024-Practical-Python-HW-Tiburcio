````python
## import libraries
import pandas as pd
import glob
````

````python
# glob everything into a list, target only the .txt files
target_files = sorted(glob.glob("project-docs/*.txt"))
target_files
````

````python
## target year, cognition_related , medical_condition, care_hours
## Line 0 is year, line 2 is medical condtion, line 4 is care hours
# Whether or not it's cognition-related is determined by the presence of "not" in line 1

for target_file in target_files:
    with open (target_file, "r") as my_text:
        print(f"Source file: {target_file}")
        all_text = my_text.readlines()
        print(f"Year: {all_text[0]}")
        print(f"Condtion: {all_text[2]}")
        print(f"Care Hours: {all_text[4]}")
        print(f"***********")
        
````

````python
## target year, cognition_related , medical_condition, care_hours
# Whether or not it's cognition-related is determined by the presence of "not" in line 2
## In the cognition_related column, enter True if the condition is related to Dementia or Alzheimer's disease. False if it is not.
## In the medical_conditions column, enter either “Dementia" or “Alzheimer’s" or “Not Specified” depending on the case.
## In the care_hours column, enter either "half-day" for 12-hour care, "full-day" for 24-hour care or “Not Specified"

conditions_list = []
for target_file in target_files:
    with open (target_file, "r") as my_text:
        all_text = my_text.readlines()
        print(f"Source file: {target_file}")
        year = all_text[0].replace("Year: ", "").rstrip()
        print(year)
        
        medical_condition = all_text[2]
        if "Dementia" in medical_condition:
            medical_conditon = "Dementia"
        elif "Alzheimer's" in medical_condition:
            medical_condition = "Alzheimer's"
        else:
            medical_condition = "Not Specified"
        print(medical_condition)
        
        cognition_related = all_text[2]
        if "not" in medical_condition:
            cognition_related = False
        else:
            cognition_related = True
        print(cognition_related)
        
        care_hours = all_text[4]
        if "12" in care_hours:
            care_hours: "half-day"
        elif "24" in care_hours:
            care_hours = "full-day"
        else:
            care_hours = "Not Specified"
        print(care_hours)
        
        conditions_list.append({"year":year,
                                "care_hours":care_hours,
                                "cognition_related":cognition_related,
                                "medical_condition":medical_condition,
                                "source_file":target_file})
        
        print(f"***********")

````
